## Import the relevant libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'Categorical.csv'</i>.

```python
data = pd.read_csv('Categorical.csv')
data
```

Remove the duplicate index column from the dataset. 

## Map the data

Use the <i>'continent'</i> category for this analysis.

```python
data_mapped = data.copy()
data_mapped['continent'] = data_mapped['continent'].map({'North America':0,'South America':1,'Europe':2,'Oceania':3,'Asia':4,'Africa':5,'Australia':6})
data_mapped
```

## Select the features

```python
work_data = data_mapped.iloc[:,3:4]
work_data
```

## Clustering

Use 4 clusters initially.

```python
kmeans = KMeans(4)
```

## Clustering results

```python
identified_clusters = kmeans.fit_predict(work_data)
identified_clusters
```
