## Import the relevant libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'Countries.csv'</i>.


```python
data = pd.read_csv('Countries-exercise.csv')
data
```

## Plot the data

Plot the <i>'Longtitude'</i> and <i>'Latitude'</i> columns. 

```python
plt.scatter(data['Longitude'],data['Latitude'])
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show()
```

## Select the features

Create a copy of that data and remove all parameters apart from <i>Longitude</i> and <i>Latitude</i>.

```python
work_data = data.drop(['name'],axis=1)
work_data
```

## Clustering

Assume there are only two clusters. 

```python
kmeans = KMeans(2)
```

```python
kmeans.fit(work_data)
```

### Clustering Resutls

```python
identified_clusters = kmeans.fit_predict(work_data)
identified_clusters
```

```python
data_w_clusters = data.copy()
data_w_clusters['Clusters'] = identified_clusters
data_w_clusters
```

Did you remember to use the <i> c </i> parameter to separate the data by the clusters we defined?

```python
plt.scatter(data_w_clusters['Longitude'],data_w_clusters['Latitude'],c=data_w_clusters['Clusters'],cmap='rainbow')
plt.show()
```
