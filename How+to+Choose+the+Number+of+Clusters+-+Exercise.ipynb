## Import the relevant libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'Countries_exercise.csv'</i>.

```python
data = pd.read_csv('Countries_exercise.csv')
data
```

Remove the duplicate index column from the dataset. 

Plot the <i>'Longtitude'</i> and <i>'Latitude'</i> columns. 

## Plot the data

Plot the <i>'Longtitude'</i> and <i>'Latitude'</i> columns. 

```python
data2 = data.copy()
plt.scatter(data2['Longitude'], data2['Latitude'])
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show()
```

## Select the features

Make sure to select the appropriate features since we are no longer using the categorical variable for our clustering but rather <i>'Longitude'</i> and <i>'Laditude'</i>.

```python
data2 = data2.drop(['name'],axis=1)
data2
```

## Clustering

Use 4 clusters initially.

```python
kmeans = KMeans(4)
kmeans.fit(data2)
```

### Clustering results

```python
identified_clusters = kmeans.fit_predict(data2)
data2['Clusters'] = identified_clusters
data2
```

## Plot the data

```python
plt.scatter(data2['Longitude'],data2['Latitude'],c=data2['Clusters'],cmap='rainbow')
plt.show()
```

## Selecting the number of clusters

### WCSS

Use the ingerated <i>sklearn</i> method <i> 'inertia_' </i>.

```python
kmeans.inertia_
```

Write a loop that calculates and saves the WCSS for any number of clusters from 1 up to 10 (or more if you wish).

```python
wcss = []
for i in range (1,50):
    kmeans = KMeans(i)
    kmeans.fit(data2)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
```
